```python
import sqlite3
import json

# JSON 파일과 SQLite DB 파일 경로
json_file = "./app/reviews.json"
db_file = "hw4.db"

# SQLite 데이터베이스 연결
conn = sqlite3.connect(db_file)
cursor = conn.cursor()

# JSON 데이터 읽기
with open(json_file, "r", encoding="utf-8") as f:
    reviews = json.load(f)

# JSON 데이터를 SQLite에 삽입
for review in reviews:
    try:
        cursor.execute("""
        INSERT INTO Review (review_id, rec_id, author, content)
        VALUES (?, ?, ?, ?)
        """, (
            review["review_id"],
            review["rec_id"],
            review["author"],
            review["review"]
        ))
    except sqlite3.IntegrityError as e:
        print(f"Skipping duplicate entry: {e}")

# 변경 사항 저장 및 연결 닫기
conn.commit()
conn.close()

print("Data successfully inserted into SQLite!")
```
